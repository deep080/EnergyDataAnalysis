```python
import pandas as pd
import numpy as np

june = np.arange(20220601,20220631)
july = np.arange(20220701,20220732)

merge = []

for date in june:
    import warnings
    warnings.simplefilter("ignore")
    df = pd.read_excel(str(date)+'.xlsx',0)
    df.drop([0,1,2,29],inplace=True)
    df.head()
    df.reset_index(inplace = True, drop = True)
    df.columns = df.iloc[0]
    df['date'] = str(date)[:4]+'-'+str(date)[4:6]+'-'+str(date)[6:]
    df.drop([0],inplace=True)
    df.reset_index(inplace = True, drop = True)
    df.drop([24],inplace=True)
    df['DateTime'] = pd.to_datetime(df['date']) + pd.to_timedelta(df['Market Hour Ending'],unit='hour')
    df.insert(0, 'DateTime', df.pop('DateTime'))
    merge.append(df)
    
for date in july:
    import warnings
    warnings.simplefilter("ignore")
    df = pd.read_excel(str(date)+'.xlsx',0)
    df.drop([0,1,2,29],inplace=True)
    df.head()
    df.reset_index(inplace = True, drop = True)
    df.columns = df.iloc[0]
    df['date'] = str(date)[:4]+'-'+str(date)[4:6]+'-'+str(date)[6:]
    df.drop([0],inplace=True)
    df.reset_index(inplace = True, drop = True)
    df.drop([24],inplace=True)
    df['DateTime'] = pd.to_datetime(df['date']) + pd.to_timedelta(df['Market Hour Ending'],unit='hour')
    df.insert(0, 'DateTime', df.pop('DateTime'))
    merge.append(df)
    
df_final = pd.concat(merge)
df_final.reset_index(inplace = True, drop = True)
#df_final.drop(df_final.columns[np.arange(27,39)],axis=1,inplace=True)
#header = pd.MultiIndex.from_product([['Central','North','South'],
                                     #['Coal','Gas','Nuclear','Hydro','Wind','Solar','Other']],
                                    #names=['Region','Fuel'])
#df.columns=header
df_final.to_excel('merged_data.xlsx')
```
